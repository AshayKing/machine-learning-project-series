```python
import numpy as np
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### Distribution Plot

```python
sns.distplot(tips['total_bill'])
```

```python
 sns.distplot(tips['tip'], rug=True, hist=False)
```

```python
from scipy.stats import norm
sns.distplot(tips['tip'], fit=norm, kde=False, vertical=True)
```

### KDE plot

```python
sns.kdeplot(tips['total_bill'], shade=True, color="r")
```

```python
sns.kdeplot(tips['total_bill'], tips['tip'])
```

```python
lunnch = tips[tips.time=='Lunch']
dinner = tips[tips.time=='Dinner']

sns.kdeplot(lunnch['total_bill'], lunnch['tip'])
sns.kdeplot(dinner['total_bill'], dinner['tip'])
```
