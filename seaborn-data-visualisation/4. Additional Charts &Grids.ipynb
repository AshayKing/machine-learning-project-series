```python
import numpy as np
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

### Joint plots

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg")
```

```python
sns.jointplot("total_bill", "tip", data=tips, kind="hex")
```

```python
sns.jointplot("total_bill", "tip", data=tips, kind="kde")
```

### Heatmap

```python
sns.heatmap(tips.corr(), annot = True, cmap= 'coolwarm', linewidths=3, linecolor='black')
```

```python
matrix = np.triu(tips.corr()) # np.tril
sns.heatmap(tips.corr(), annot=True, mask=matrix)
```

### Pair Plot

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
```

```python
sns.pairplot(iris, vars=["sepal_width", "sepal_length"])
```

```python
sns.pairplot(iris, diag_kind="kde", kind="reg")
```

### FacetGrid

```python
fg = sns.FacetGrid(tips, col="time", row="smoker")

bins = np.arange(0, 65, 5)
fg.map(plt.hist, "total_bill", bins=bins, color="r")
```

```python
fg = sns.FacetGrid(tips, col="time",  row="smoker")
fg.map(plt.scatter, "total_bill", "tip", edgecolor="w")
```
