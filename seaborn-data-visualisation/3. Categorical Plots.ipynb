```python
import numpy as np
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### Strip Plot

```python
sns.stripplot(x="sex", y="total_bill", hue="day", data=tips, palette='Set3',dodge=True)
```

```python
sns.stripplot(x="total_bill", y="day", data=tips, linewidth=1)
```

### Swarm Plot

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill", hue="smoker",
                   data=tips, palette="Set2", dodge=True)
```

### Box Plot

```python
sns.boxplot(x=tips["total_bill"])
```

```python
sns.boxplot(x="day", y="total_bill", hue='smoker', data=tips, palette='Set2', linewidth=2.5)
```

```python
sns.boxplot(data=tips, orient="h", palette="Set2")
```

### Violin Plot

```python
sns.violinplot(x="day", y="total_bill", hue="smoker", data=tips, split=True)
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex",
                    data=tips, palette="Set2",
                    scale="count", inner="quartile")
```

### Point Plot

```python
sns.pointplot(x="time", y="total_bill", hue="smoker",
                   data=tips, markers=["o", "x"],linestyles=["-", "--"])
```

### Bar Plot

```python
sns.barplot(x="day", y="total_bill", data=tips)
```

```python
sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
```

```python
from numpy import median
sns.barplot(x="day", y="tip", data=tips, estimator=median)
```

### Count Plot

```python
sns.countplot(x="day", data=tips)
```

```python
sns.countplot(y="day", hue="time", data=tips)
```

### Cat Plot 

* In case we want to plot any of above plot with multiple cat features, we have to use catply
* we have provide type are any of above plot

```python
sns.catplot(x="time", hue="day", col="smoker",
                data=tips, kind="count",
                height=4, aspect=.7);
```

```python
sns.catplot(x="sex", y="total_bill",
                hue="smoker", col="time",
                data=tips, kind="bar",
                height=4, aspect=.7);
```

### Two plots in one dig

```python
sns.boxplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
```
