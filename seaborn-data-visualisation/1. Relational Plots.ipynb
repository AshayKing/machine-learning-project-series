```python
import numpy as np
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### Relational Plots

```python
# Plotting releation between two numerical features with hue are categorical feature
sns.relplot(x="total_bill", y="tip", hue="day", data=tips)
```

```python
# Along with above mentioned arguments we can specify row & col arguments (specify categorical features)
# it creates #(unique cat of row feature)X#(unique cat of col feature) for every combination.

sns.relplot(x="total_bill", y="tip", hue="day", row='time', col='sex', data=tips)
```

```python
sns.relplot(x="total_bill", y="tip",
                hue="time", style="time", col="day",
                height=5, aspect=.7,
                kind="line", data=tips)
```

### Scatter Plots

```python
# Sample like relplot
sns.scatterplot(x="total_bill", y="tip", hue="time", size="size", sizes=(20, 200), palette='coolwarm', data=tips)
```

```python
markers = {"Lunch": "s", "Dinner": "X"}
sns.scatterplot(x="total_bill", y="tip", style="time",
                     markers=markers,
                     data=tips)
```
